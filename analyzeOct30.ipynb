```python
import pandas as pd
import numpy as np
headers = ('a','b','c','d')
df1 = pd.read_csv('CMS_PIX28_OCT30_v2_SINGLE_PIX_ALL.csv')
df1.columns=headers
df1
```

```python
df1.drop('a', axis=1, inplace=True)
```

```python
df1[0:15]
```

```python
#df1 = df1.astype('uint')
#df1
#print(df1.dtypes)
```

```python
df1['converted'] = df1.d.apply(lambda x: format(int(x), '08b'))
df1
```

```python
df1['firstBit']=df1.converted.apply(lambda x: x[1:2])
df1
```

```python
df1['secondBit']=df1.converted.apply(lambda x: x[2:3])
df1
```

```python
df1['thirdBit']=df1.converted.apply(lambda x: x[3:4])
df1
```

```python
df1.to_csv('converted_Oct30.csv')
```

```python
df2 = pd.read_csv('converted_Oct30.csv')
df2.drop('Unnamed: 0', axis=1, inplace=True)
df2.drop('c', axis=1, inplace=True)
```
